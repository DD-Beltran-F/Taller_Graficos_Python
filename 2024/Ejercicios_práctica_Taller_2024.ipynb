## 1. Ejercicio 1: *춰La aldea del Reloj de Sol necesita tu ayuda!*  <a name="ej1"></a>

La **aldea del Reloj de Sol** est치 pasando por un mal momento, sus relaciones con las dem치s aldeas se est치 debilitando y sus recursos se est치n acabando 游땞.

La funci칩n que modela la cantidad de comida por habitante en funci칩n del tiempo (en a침os) es:

$$
f(t) = 0.4\cdot (t + 6)^3 \cdot e^{-0.5t}
$$

Si la aldea mejora su relaci칩n con sus vecinos la **aldea Achloclonados**, 춰entonces recibir치n nuevas provisiones de choclos cada a침o! 游꺟游꺟游꺟, si esto ocurre entonces la funci칩n que modela la cantidad de comida por habitante en funci칩n del tiempo (en a침os) es:

$$
g(t) = 5cos(4t)sin(t) + 86.4
$$

춰OH NO! La **aldea de Las Palmeras** est치 enojada con nuestra gente 游놐. El jefe Palmera dijo que si esta temporada no llov칤a, saquear칤an la *aldea del Reloj de Sol* hasta que les robaran toda la comida. Si esto ocurre, la funci칩n que modela la cantidad de comida por habitante en funci칩n del tiempo (en a침os) es:

$$
h(t) = e^{4.5 - t}
$$

El jefe de la aldea quiere estudiar el comportamiento de aqu칤 a 30 a침os (entonces consideramos $t = [0,30]$), as칤 que te pidi칩 que le entregaras una hoja con 3 gr치ficos que muestren el comportamiento de la cantidad de comida por persona a lo largo de los 30 a침os. Como ya somos expertos en utilizar *matplotlib*, crearemos 3 *subplots* dentro de una figura para que el jefe pueda leer la informaci칩n sin problemas.

#### **No olvides hacer los imports necesarios**, puedes personalizar los gr치ficos de la manera que tu quieras 游뱔

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table
import matplotlib.font_manager as fm
```

```python
font1 = {'family': 'Serif',
            'color':  'black',
            'weight': 'normal',
            'size': 20,
            }
font2 = fm.FontProperties(family='Serif',size=20,weight='normal')
```

```python
# Definimos las funciones
def f(t):
    return 0.4*(t + 6)**3 * np.exp(-0.5 * t)

def g(t):
    return 5*np.cos(4*t)*np.sin(t) + 86.4

def h(t):
    return np.exp(4.5 - t)

# Evaluamos las funciones
eje_x = np.linspace(0,30,1000) # EJE X
eje_y1 = [f(t) for t in eje_x] # FUNCI칍N 1
eje_y2 = [g(t) for t in eje_x] # FUNCI칍N 2
eje_y3 = [h(t) for t in eje_x] # FUNCI칍N 3

fig, ax = plt.subplots(nrows=3,ncols=1,figsize=(8, 24)) # Tambi칠n sirve subplots(2,2,...)
fig.subplots_adjust(hspace=0.3, wspace=0.5)

ax[0].plot(eje_x,eje_y1,color='k',lw=3) # Graficamos la funci칩n 1
ax[0].set_xlabel('Tiempo (a침os)',fontdict=font1)
ax[0].set_ylabel('Comida',fontdict=font1)
ax[0].set_title('Comida de la aldea a lo largo de 30 a침os',size=25,fontdict=font1,pad=15)
ax[0].set_xlim(0,30)
ax[0].set_ylim(0,100)
ax[0].set_xticks(ax[0].get_xticks(),ax[0].get_xticklabels(),fontdict=font1)
ax[0].set_yticks(ax[0].get_yticks(),ax[0].get_yticklabels(),fontdict=font1)
ax[0].minorticks_on()
ax[0].tick_params(direction='in',length=16,width=2,labelsize=20,which='major',right=True,top=True)
ax[0].tick_params(direction='in',length=6,width=2,labelsize=20,which='minor',right=True,top=True)
ax[0].grid()

ax[1].plot(eje_x,eje_y2,color='r',lw=3) # Graficamos la funci칩n 2
ax[1].set_xlabel('Tiempo (a침os)',fontdict=font1)
ax[1].set_ylabel('Comida',fontdict=font1)
ax[1].set_title('Comida de la aldea con ayuda',size=25,fontdict=font1,pad=15)
ax[1].set_xlim(0,30)
ax[1].set_ylim(0,100)
ax[1].set_xticks(ax[1].get_xticks(),ax[1].get_xticklabels(),fontdict=font1)
ax[1].set_yticks(ax[1].get_yticks(),ax[1].get_yticklabels(),fontdict=font1)
ax[1].minorticks_on()
ax[1].tick_params(direction='in',length=16,width=2,labelsize=20,which='major',right=True,top=True)
ax[1].tick_params(direction='in',length=6,width=2,labelsize=20,which='minor',right=True,top=True)
ax[1].grid()

ax[2].plot(eje_x,eje_y3,color='green',lw=3) # Graficamos la funci칩n 3
ax[2].set_xlabel('Tiempo (a침os)',fontdict=font1)
ax[2].set_ylabel('Comida',fontdict=font1)
ax[2].set_title('Comida de la aldea bajo ataque',size=25,fontdict=font1,pad=15)
ax[2].set_xlim(0,30)
ax[2].set_ylim(0,100)
ax[2].set_xticks(ax[2].get_xticks(),ax[2].get_xticklabels(),fontdict=font1)
ax[2].set_yticks(ax[2].get_yticks(),ax[2].get_yticklabels(),fontdict=font1)
ax[2].minorticks_on()
ax[2].tick_params(direction='in',length=16,width=2,labelsize=20,which='major',right=True,top=True)
ax[2].tick_params(direction='in',length=6,width=2,labelsize=20,which='minor',right=True,top=True)
ax[2].grid()

plt.show() # Muestra la figura
```

## 2. Ejercicio 2: *춰쯃os m치s r치pidos del ombligo?!*  <a name="ej2"></a>

Un grupo de estudiantes de astronom칤a decidieron medir el tiempo de reacci칩n de sus compa침eros en el ombligo 游땓. La mayor칤a estaba bien despierto, sin embargo algunos reaccionaban un pooooooquito lento 游낿.

Lee los datos en el archivo $tiempos\_de\_reaccion.csv$ y guardalos en una tabla de *astropy* 游늵.

1. Grafica los datos en un histrograma con el n칰mero de bines adecuados dada la cantidad de datos.

2. Calcula la media $\mu$ y desviaci칩n est치ndar $\sigma$ de los datos *(hint: busca la funci칩n $mean()$ y $std()$ de numpy)*.

3. Crea otro gr치fico donde muestres el histograma, la media $\mu$ representada por una *l칤nea vertical*, y dos l칤neas verticales para $\mu - \sigma$ y $\mu + \sigma$. Procura asignarle estilos defiretenes a las l칤neas para poder diferenciarlas.

3. En el mismo gr치fico del punto $3.$, a침ade $labels$ a las l칤neas verticales y muestralos en una leyenda.

#### **No olvides hacer los imports necesarios**, puedes personalizar los gr치ficos de la manera que tu quieras 游뱔

```python
# Leemos los datos
tabla = Table.read('tiempos_de_reaccion.csv') # Leemos el archivo en la ruta "funciones.csv"
tabla
```

```python
tiempos = tabla['datos']
```

```python
fig,ax = plt.subplots(figsize=(10, 10))
ax.hist(tiempos,bins=32,color='deeppink',lw=2) #Crea un histograma con 10 bins
ax.set_xlabel('Tiempo (s)',fontdict=font1)
ax.set_ylabel('Frecuencia',fontdict=font1)
ax.set_title('Tiempos de reacci칩n de los del ombligo',size=25,fontdict=font1,pad=15)
ax.set_xticks(ax.get_xticks(),ax.get_xticklabels(),fontdict=font1)
ax.set_yticks(ax.get_yticks(),ax.get_yticklabels(),fontdict=font1)
ax.tick_params(direction='inout',length=16,width=1.5,labelsize=20)
plt.show()
```

```python
# Calculamos la media
media = np.mean(tiempos)

# Calculamos la desviaci칩n est치ndar
std = np.std(tiempos)

print(f'Media: {media}')
print(f'Desviaci칩n est치ndar: {std}')
```

```python
# Graficamos el histograma junto a las l칤neas verticales
fig,ax = plt.subplots(figsize=(10, 10))
ax.hist(tiempos,bins=32,color='deeppink',lw=2) #Crea un histograma con 10 bins
ax.set_xlabel('Tiempo (s)',fontdict=font1)
ax.set_ylabel('Frecuencia',fontdict=font1)
ax.set_title('Tiempos de reacci칩n de los del ombligo',size=25,fontdict=font1,pad=15)
ax.set_xticks(ax.get_xticks(),ax.get_xticklabels(),fontdict=font1)
ax.set_yticks(ax.get_yticks(),ax.get_yticklabels(),fontdict=font1)
ax.tick_params(direction='inout',length=16,width=1.5,labelsize=20)

ax.axvline(media,c='blue',lw=3,label=r'$\mu$')
ax.axvline(media + std,c='k',ls=':',lw=3,label=r'$\mu + \sigma$')
ax.axvline(media - std,c='k',ls=':',lw=3,label=r'$\mu - \sigma$')

ax.legend(loc='best',prop=font2)

plt.show()
```
